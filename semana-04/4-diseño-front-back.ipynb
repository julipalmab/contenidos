## Dise침o de software: *Front-end* y *Back-end*

En la ingenier칤a de software existen los conceptos de ***front-end*** y ***back-end*** para referirse a la separaci칩n que existe entre la capa de presentaci칩n y la capa de acceso a los datos, respectivamente. 

En el caso de interfaces gr치ficas:

- el ***front-end*** est치 relacionado a la **interfaz gr치fica** con la cual el usuario interact칰a, y 
- el ***back-end*** se refiere a la **l칩gica** detr치s de ella. 

Esta separaci칩n se alinea con un principio importante en el dise침o de *software* de calidad que indica que siempre debemos buscar **alta cohesi칩n y bajo acomplamiento** en nuestros dise침os.

- **Cohesi칩n**: Cada una de las componentes del software debe realizar **solo las tareas para las cuales fue creada**, delegando otras tareas a otras componentes seg칰n corresponda. Por ejemplo, si tengo una clase `Simulaci칩nDeParque`, un dise침o altamente cohesionado incluir칤a m칠todos como `iniciar_simulaci칩n()` o `detener_simulaci칩n()`, pero no m칠todos como `limpiar_atracci칩n()` o `ingresar_clientes_a_restaurant()`, ya que la clase `Simulaci칩nDeParque` fue dise침ada para administrar la simulaci칩n y no para hacerse cargo de m칠todos que deber칤an ser ejecutados por (*delegados a*) otras clases de la simulaci칩n.

- **Acoplamiento**: Cuando la modificaci칩n de una componente implica que es necesario modificar otra componente para que la implementaci칩n del cambio sea correcta y completa. Por ejemplo, si al modificar los atributos de una clase A, tambi칠n se deben modificar los atributos de otra clase B, se dice que hay *alto acoplamiento* entre las clases A y B. Un buen dise침o intenta reducir el acoplamiento entre clases.

De ah칤 que siempre debemos buscar **ALTA COHESI칍N y BAJO ACOPLAMIENTO** en nuestros dise침os.

Escribir c칩digo con separaci칩n del ***front-end*** y ***back-end***, significa implementar las funciones de un programa con una clara separaci칩n entre entidades dedicadas a lo visual (***front-end***) y entidades dedicadas a detalles algoritmicos, de datos, o de l칩gica (***back-end***). Algunas ventajas de hacerlo son las siguientes:

1. **Modularidad**: Permite cambiar cualquiera de las dos partes sin afectar la otra (bajo acoplamiento). En particular, podemos editar el *front-end* suponiendo que las funciones utilizadas por el *back-end* mantienen su comportamiento. Al mismo tiempo, es posible modularizar el *back-end* de manera independiente del *front-end*. Podemos reescribir el c칩digo para hacerlo cada vez m치s eficiente y espec칤fico (alta cohesi칩n). Podemos incluso modularizar el *back-end* en muchos archivos distintos y luego consultar todas las funcionalidades desde un solo archivo de conexi칩n con el *front-end* (alta cohesi칩n).
1. **Uso de recursos**: Algunas veces el *front-end* est치 corriendo en un computador distinto al *back-end*. Si el *back-end* ejecuta c치lculos muy costosos, no nos gustar칤a cargarle este tiempo computacional a la interfaz gr치fica. Un ejemplo claro de esto son los navegadores de internet (*browsers*), donde la mayor칤a de los c치lculos o datos que queremos obtener se generan en un servidor de *back-end* y el resultado obtenido solo se muestra gr치ficamente en el computador del usuario (*front-end*). De este modo, nuestro computador no tiene que sobrecargarse procesando cosas.
1. **Escalamiento**: Por un lado, permite hacer crecer un *software* sin mucha interferencia a las funcionalidades antiguas. Por otro lado, permite distribuir el procesamiento en el *back-end* utilizando m칰ltiples r칠plicas de este, lo que es muy usado en la web.
1. **Experticia**: Usualmente los desarrollador de un *front-end* tienen un tipo de experiencia muy distinta a la que tienen quienes desarrollan el *back-end*. Mantenerlos separados permite obtener lo mejor de ambas partes.
1. **Mantenci칩n**: Es posible hacer *testing* parte por parte de una pieza de software, e introducir correcciones o mejoras evitando que un alto acoplamiento de funcionalidades haga que las modificaciones deban ser propagadas a m칰ltiples partes del c칩digo.
1. **Evoluci칩n del software/versionamiento**: Si quieres cambiar completamente una de las partes puedes hacerlo sin problema, mientras las funciones utilizadas en el *front-end* sigan teniendo los mismos nombres que antes (dicho de otra forma, mientras se mantenga la misma interfaz de m칠todos). De esta manera, por ejemplo, si programas un *back-end* para PyQt5 y luego quieres usarlo para PyQt6, puedes hacerlo sin problemas (alta cohesi칩n). O mejor a칰n, exportar tus funcionalidades para un *back-end* web.

Estas ventajas pueden transformarse en costos si es que hay que tener dos equipos distintos de desarrollo o hay que mantener dos *software* distintos. Sin embargo, las ventajas siguen siendo mayores.

## Ejemplo: Escribiendo un programa separando *front-end* y *back-end*

Se mostrar치 mediante ejemplos la evoluci칩n de un programa que inicialmente no sigue la arquitectura *front-end*/*back-end*, y termina sigui칠ndola. El programa que se construye tiene por objetivo recibir una lista de n칰meros (separados por coma), y muestra como resultado la lista ordenada de n칰meros.

### Versi칩n 1: Programa poco cohesivo

El siguiente programa resuelve lo pedido, y sigue el siguiente flujo:

- Define los componentes de interfaz gr치fica: dos *labels*, un *input* de texto y un bot칩n.
- Conecta el clic sobre el bot칩n con la funcionalidad de ordenamiento.
- Al hacer clic y gatillar el evento de ordenar:
    - Por simplicidad, se remueven todos los espacios de `texto_input` y comas sobrantes.
    - Se verifica que el texto recibido (la lista de n칰meros ingresada por el usuario) sea v치lido
    - Si no es v치lido, se notifica al usuario mediante una etiqueta de texto.
    - Si lo es, se ordenan los n칰meros mediante un algoritmo que saca el m칤nimo sucesivamente y luego muestra el resultado.
    
Este c칩digo se encuentra en el archivo `dise침o_front_back_version_1.py`.

```python
import sys
from PyQt6.QtWidgets import (QApplication, QWidget, QLabel, QPushButton, QLineEdit)


class Ventana(QWidget):

    def __init__(self):
        super().__init__()
        self.inicializa_gui()

    def inicializa_gui(self):

        self.etiqueta = QLabel(
            'Ingresa una lista de n칰meros separados por comas:', self)
        self.etiqueta.move(20, 10)
        self.etiqueta.resize(self.etiqueta.sizeHint())

        self.input = QLineEdit('', self)
        self.input.setGeometry(20, 40, 360, 20)

        self.boton = QPushButton('Ordenar', self)
        self.boton.setGeometry(20, 70, 360, 30)
        self.boton.clicked.connect(self.revisar_y_ordenar)

        self.resultado = QLabel('', self)
        self.resultado.move(20, 100)
        self.resultado.resize(self.resultado.sizeHint())

        self.setGeometry(700, 300, 400, 200)
        self.setWindowTitle('Ordenador de n칰meros')
        self.show()

    def revisar_y_ordenar(self):
        texto_input = self.input.text()
        if texto_input:
            texto_input = texto_input.replace(' ', '').strip(',')
            caracteres_posibles = list(map(str, range(0, 10))) + [',']
            for caracter in texto_input:
                if caracter not in caracteres_posibles:
                    self.resultado.setText('Input no v치lido')
                    self.resultado.resize(self.resultado.sizeHint())
                    return
            lista_de_numeros = [int(porcion)
                                for porcion in texto_input.split(',')]
            numeros_ordenados = []
            while len(lista_de_numeros) > 0:
                minimo_actual = lista_de_numeros[0]
                for numero in lista_de_numeros:
                    if numero < minimo_actual:
                        minimo_actual = numero
                numeros_ordenados.append(minimo_actual)
                lista_de_numeros.remove(minimo_actual)
            texto_resultado = ", ".join([str(numero)
                                        for numero in numeros_ordenados])
            self.resultado.setText(texto_resultado)
            self.resultado.resize(self.resultado.sizeHint())
            self.resultado.repaint()
```

Si bien el c칩digo logra lo esperado, notamos primero que es muy poco **cohesivo**. La clase `Ventana` realiza **todas** las operaciones necesarias para lograr el resultado del programa. La clase `Ventana` al ser un componente visual, solo debiese encargarse de tareas visuales que involucren a esta ventana. El m칠todo `revisar_y_ordenar` espec칤ficamente realiza tareas de procesamiento de datos **y** tareas visuales en la ventana. Las responsabilidades de tareas no se reparten correctamente entre m치s componentes, por lo que resulta en c칩digo poco cohesivo.

### Versi칩n 2: Programa cohesivo pero acoplado

En la siguiente versi칩n del programa se logra separar las responsabilidades de c칩digo en distintos componentes l칩gicos. Se crea un m칩dulo llamado `dise침o_front_back_version_2_backend.py` que se encarga de la validaci칩n del *input* y ordenamiento del la lista de n칰meros, mientras que el m칩dulo `dise침o_front_back_version_2_frontend.py` se encarga del aspecto visual.

Pero, 쯖칩mo se soluciona esta situaci칩n? **Hay** que llamar de una forma u otra a esta funci칩n para lograr el objetivo resultante, 쯖칩mo se logra sin llamar a la funci칩n directamente? 춰Se침ales! 

### Versi칩n 3: Programa cohesivo y poco acoplado

En esta 칰ltima versi칩n, se logra una separaci칩n de componentes *front-end*/*back-end* que logra un programa altamente cohesivo, y mediante comunicaci칩n via se침ales se logra un programa levemente acoplado.

La diferencia con la 칰ltima versi칩n es que se elimina el llamado directo de objetos de distintos componentes y se delega todo a se침ales para comunicarse. Se crean dos se침ales: `senal_actualizar` cuyo objetivo es comunicar desde *back-end* al *front-end* una actualizaci칩n de ventana, y `senal_procesar` cuyo objetivo es comunicar desde *front-end* a *back-end* cuando es necesario procesar *input* nuevo.

Ahora las funciones de procesamiento se encapsulan en una clase `Procesador` para contener la se침al de PyQt, y en el c칩digo principal es necesario conectar las se침ales de una clase con el m칠todo correspondiente de la otra clase. En este caso:
* La `senal_procesar` del *front-end* se conecta con el m칠todo `procesar_input` del *back-end*.
* La `senal_actualizar` del *back-end* se conecta con el m칠todo `actualizar_resultado` del *front-end*.

Este 칰ltimo detalle es necesario para que efectivamente se comuniquen los componentes, y es inevitablemente un aspecto de acoplamiento del programa. Es imposible lograr acoplamiento cero en un programa, pero si es posible **minimizarlo**.

**M칩dulo: dise침o_front_back_version_3_frontend.py**

```python
import sys
from PyQt6.QtCore import pyqtSignal
from PyQt6.QtWidgets import (QApplication, QWidget, QLabel, QPushButton, QLineEdit)
from procesador import Procesador


class Ventana(QWidget):

    senal_procesar = pyqtSignal(str)

    def __init__(self):
        super().__init__()
        self.inicializa_gui()

    def inicializa_gui(self):

        self.etiqueta = QLabel('Ingresa una lista de n칰meros separados por comas:', self)
        self.etiqueta.move(20, 10)
        self.etiqueta.resize(self.etiqueta.sizeHint())

        self.input = QLineEdit('', self)
        self.input.setGeometry(20, 40, 360, 20)

        self.boton = QPushButton('Ordenar', self)
        self.boton.setGeometry(20, 70, 360, 30)
        self.boton.clicked.connect(self.boton_clickeado)

        self.resultado = QLabel('', self)
        self.resultado.move(20, 100)
        self.resultado.resize(self.resultado.sizeHint())

        self.setGeometry(700, 300, 400, 150)
        self.setWindowTitle('Ordenador de n칰meros')
        self.show()

    def boton_clickeado(self):
        texto_input = self.input.text()
        self.senal_procesar.emit(texto_input)

    def actualizar_resultado(self, texto):
        self.resultado.setText(texto)
        self.resultado.resize(self.resultado.sizeHint())
        self.resultado.repaint()


if __name__ == '__main__':
    app = QApplication([])
    procesador = Procesador()
    ventana = Ventana()
    procesador.senal_actualizar.connect(ventana.actualizar_resultado)
    ventana.senal_procesar.connect(procesador.procesar_input)
    sys.exit(app.exec())
    
```

**M칩dulo: dise침o_front_back_version_3_backend.py**
```python
from PyQt6.QtCore import QObject, pyqtSignal


class Procesador(QObject):

    senal_actualizar = pyqtSignal(str)

    def es_valido(self, texto):
        caracteres_posibles = list(map(str, range(0, 10))) + [',']
        for caracter in texto:
            if caracter not in caracteres_posibles:
                return False
        return True

    def ordenar(self, lista_de_numeros):
        numeros_ordenados = []
        while len(lista_de_numeros) > 0:
            minimo_actual = lista_de_numeros[0]
            for numero in lista_de_numeros:
                if numero < minimo_actual:
                    minimo_actual = numero
            numeros_ordenados.append(minimo_actual)
            lista_de_numeros.remove(minimo_actual)
        return numeros_ordenados

    def procesar_input(self, texto_input):
        texto_input = texto_input.replace(' ', '').strip(',')
        if not self.es_valido(texto_input):
            self.actualizar_interfaz('Input no v치lido')
            return
        lista_de_numeros = [int(porcion) for porcion in texto_input.split(',')]
        numeros_ordenados = self.ordenar(lista_de_numeros)
        texto_resultado = ", ".join([str(numero)
                                    for numero in numeros_ordenados])
        self.actualizar_interfaz(texto_resultado)

    def actualizar_interfaz(self, texto):
        self.senal_actualizar.emit(texto)
```

Es importante notar que el c칩digo dentro de las dos clases principales creadas (`Ventana` y `Procesador`) nunca llama directamente al c칩digo de la otra clase, solo llama y ejecuta c칩digo interno. Cada una funciona por si sola y delega la responsabilidad de envio de informaci칩n a se침ales.

En general, para escribir un programa de:
- _Software_ altamente **cohesivo**, la estrategia es separar de forma adecuada las **responsabilidades** de los distintos compontentes del programa; y escribir distintos m칩dulos y clases destintados a objetivos claros.

- _Software_ poco **acoplado**, la estrategia es independizar lo m치s posible los distintos compontentes; el uso de **se침ales** es muy compatible con esta idea ya que reduce el acoplamiento al uso de se침ales comunes.

Algunas dudas frecuentes que pueden aparecer al ver este ejemplo:

- 游뱂 쯇or qu칠 `Procesador` hereda de `QObject`? 游뱂 

Porque por dise침o se genera la se침al de procesamiento en esta clase, y por reglas de PyQt, toda clase que crea como atributo una `pyqtSignal`, es necesario que herede de `QObject` y llame a su constructor (`super().__init__()`).

- 游뱂 쯇or qu칠 una se침al se crea en *back-end* y otra se침al en *front-end*? 游뱂 

Esa fue una decisi칩n de dise침o espec칤fica de este ejemplo. Se eligi칩 crear la se침al seg칰n qui칠n emite dicha se침al: actualizar el resultado en *back-end* y procesar input en *front-end*. Es posible definir de otra forma perfectamente v치lida, pero las conexiones se han de hacer en otro orden.

- 游뱂 쯅o es sobre complicado para el ejemplo buscado? 游뱂 

Tal vez un poco, pero como se menciona anteriormente, esto es m치s aparente por que el programa es relativamente peque침o y simple. En ejemplos m치s extensos y complicados, hacer este tipo de separaci칩n es mucho m치s evidente y provechosa.
